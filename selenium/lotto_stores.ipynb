```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select, WebDriverWait
from selenium.webdriver.support.expected_conditions import visibility_of_element_located
from selenium.webdriver.support.expected_conditions import visibility_of_all_elements_located
from selenium.webdriver.support.expected_conditions import presence_of_element_located
from selenium.webdriver.support.expected_conditions import presence_of_all_elements_located

import time
import os
```

```python
with webdriver.Chrome() as driver:
    wait = WebDriverWait(driver, 10)
    driver.get("https://dhlottery.co.kr/store.do?method=sellerInfo645")
    
    '''
    # select available store only
    salesonly_opt = wait.until(presence_of_element_located((By.NAME, "rtlrSttus")))
    select = Select(salesonly_opt)
    select.select_by_value('001') # 판매점
    '''

    # preassigned search_btn
    search_btn = driver.find_element(By.ID, 'searchBtn1')
    
    for selected_num, (selected_sido, selected_gugun) in enumerate(sido_gugun_list):
        if selected_num < last_saved_num:
            continue

        # select sido, sltSIDO
        combo_sido = wait.until(presence_of_element_located((By.NAME, "sltSIDO")))
        select_sido = Select(combo_sido)
        select_sido.select_by_value(selected_sido)

        # select gugun, sltGUGUN, 구/군
        if selected_sido == '세종':
            pass
        else:
            #sltSIDO, 시/도
            combo_gugun = wait.until(presence_of_element_located((By.NAME, "sltGUGUN")))
            select_gugun = Select(combo_gugun)
            select_gugun.select_by_value(selected_gugun)

        search_btn.click()
        time.sleep(1)
        
        # Call twice for stability, Why not? I don't know exactly
        #pages = wait.until(visibility_of_all_elements_located((By.XPATH, "//div[@id='pagingView']//a")))
        #pages = wait.until(visibility_of_all_elements_located((By.XPATH, "//div[@id='pagingView']//a")))
        pages = wait.until(visibility_of_all_elements_located((By.XPATH, "//div[@id='pagingView']//a")))
        pages = wait.until(visibility_of_all_elements_located((By.XPATH, "//div[@id='pagingView']//a")))
        print(selected_gugun, len(pages))
        num_of_pages = len(pages)
        
        if num_of_pages > 10:
            num_of_pages = num_of_pages -1 # 부천시 등

        for i in range(num_of_pages):
            #page_query = "//div[@id='pagingView']//a[contains(.,'{}')]".format(i+1)
            page_query = f"//*[@id=\"pagingView\"]/a[{i+1}]"
            #print(f'i+1:{i+1}, page_query:{page_query}')

            # Not to assign page_link object from pages object directly, that makes odd behavior
            # 경기>부천시 10page 넘어감
            if i>= 10 and i%10==0:
                #next_page = wait.until(visibility_of_element_located((By.CLASS_NAME, 'go.next')))
                next_page = driver.find_element(By.CLASS_NAME, 'go.next')
                next_page.click()
                time.sleep(1)
            elif i>0:
                #page_link = wait.until(visibility_of_element_located((By.XPATH, page_query)))
                page_link = driver.find_element(By.XPATH, page_query)
                page_link.click()
                time.sleep(1)

            #rows = wait.until(visibility_of_all_elements_located((By.XPATH, "//table[@id='resultTable']//tbody//tr")))
            rows = driver.find_elements(By.XPATH, "//table[@id='resultTable']//tbody//tr")
            time.sleep(1)
            
            for row in rows:
                cols = row.find_elements(By.TAG_NAME, "td")
                time.sleep(0.1)
                store_name, store_tel, store_add = cols[0].text, cols[1].text, cols[2].text
                store_info = '{}/{}/{}/{}/{}'.format(selected_sido, \
                                                     selected_gugun, \
                                                     store_name, \
                                                     store_tel, \
                                                     store_add)
                store_list.append(store_info)
                        
        # time sleep for consitency
        time.sleep(1)
```
